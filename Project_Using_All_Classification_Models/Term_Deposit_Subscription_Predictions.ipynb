```python
job=pd.get_dummies(term_deposit['job'],prefix='job',drop_first=True,dtype='int64')
term_deposit=pd.concat([term_deposit,job],axis=1)
martial=pd.get_dummies(term_deposit['marital'],prefix='martial',drop_first=True,dtype='int64')
term_deposit=pd.concat([term_deposit,martial],axis=1)
education=pd.get_dummies(term_deposit['education'],prefix='education',drop_first=True,dtype='int64')
term_deposit=pd.concat([term_deposit,education],axis=1)
contact=pd.get_dummies(term_deposit['contact'],prefix='contact',drop_first=True,dtype='int64')
term_deposit=pd.concat([term_deposit,contact],axis=1)
month=pd.get_dummies(term_deposit['month'],prefix='month',drop_first=True,dtype='int64')
term_deposit=pd.concat([term_deposit,month],axis=1)
outcome=pd.get_dummies(term_deposit['poutcome'],prefix='poutcome',drop_first=True,dtype='int64')
term_deposit=pd.concat([term_deposit,outcome],axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=term_deposit_train.drop(['subscribed','ID'],axis=1)
y=term_deposit_train['subscribed']
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=24,random_state=42)
x_pca=pca.fit_transform(x)
print(f'Explained Variance Ratio:{pca.explained_variance_ratio_}')
print(f'Cummulative Variance Explained:{sum(pca.explained_variance_ratio_):.2f}')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier,BaggingClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,precision_recall_curve,roc_curve,roc_auc_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=42)
```

```python
lr_reg=LogisticRegression(random_state=42)
lr_reg.fit(x_train,y_train)
lr_params={'penalty':['l1','elasticnet'],'solver':['lbfgs','newton-cg','sag','saga'],'n_jobs':[1]}
lr_grid=GridSearchCV(lr_reg,param_grid=lr_params,cv=10,n_jobs=1,return_train_score=True,scoring='accuracy')
lr_grid.fit(x_train,y_train)
best_score_lr=lr_grid.best_score_
best_params_lr=lr_grid.best_params_
lr_pred=lr_grid.predict(x_test)
print(f"Classification Report:{classification_report(y_test,lr_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,lr_pred)}")
print(f"Accuracy Score:{accuracy_score(y_test,lr_pred)}")
print(f"Best Score:",best_score_lr)
print(f"Best Params:",best_params_lr)
```

```python
lr_pred_proba=lr_grid.predict_proba(x_test)[:,1]
precision,recall,thresholds=precision_recall_curve(y_test,lr_pred_proba)
plt.Figure(figsize=(8,6))
plt.plot(recall,precision,label="Logistic Regression")
plt.xlabel('recall')
plt.ylabel('Precision')
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

```python
fpr,tpr,thresholds=roc_curve(y_test,lr_pred_proba)
auc=roc_auc_score(y_test,lr_pred_proba)
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f'ROC Curve(AUC={auc:.2f})',color='blue')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
dt_clf=DecisionTreeClassifier(random_state=42)
dt_clf.fit(x_train,y_train)
dt_params={'criterion':['gini','entropy','log_loss'],'splitter':['best','random'],'max_depth':[1,3,5],'min_samples_split':[2,3],
'max_features':['auto','sqrt'] }
dt_grid=GridSearchCV(dt_clf,param_grid=dt_params,scoring='accuracy',n_jobs=-1,return_train_score=True,cv=6)
dt_grid.fit(x_train,y_train)
best_score_dt=dt_grid.best_score_
best_params_dt=dt_grid.best_params_
dt_pred=dt_grid.predict(x_test)
print(f"Classification Report:{classification_report(y_test,dt_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,dt_pred)}")
print(f"Accuracy Score:{accuracy_score(y_test,dt_pred)}")
print(f"Best Score",best_score_dt)
print(f"Best Params",best_params_dt)
```

```python
dt_pred_proba=dt_grid.predict_proba(x_test)[:,1]
precision,recall,thresholds=precision_recall_curve(y_test,dt_pred_proba)
plt.Figure(figsize=(8,6))
plt.plot(recall,precision,label="Decision Tree Classifier")
plt.xlabel('recall')
plt.ylabel('Precision')
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

```python
fpr,tpr,thresholds=roc_curve(y_test,dt_pred_proba)
auc=roc_auc_score(y_test,dt_pred_proba)
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f'ROC Curve(AUC={auc:.2f})',color='blue')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
rf_clf=RandomForestClassifier(n_estimators=100,random_state=42)
rf_clf.fit(x_train,y_train)
rf_params={'n_estimators':[100,200],'criterion':['gini','log_loss','entropy'],'max_depth':[1,3,5],'max_features':['sqrt','log2','auto']}
rf_grid=GridSearchCV(rf_clf,param_grid=rf_params,cv=6,n_jobs=-1,return_train_score=True,scoring='accuracy')
rf_grid.fit(x_train,y_train)
best_params_rf=rf_grid.best_params_
best_score_rf=rf_grid.best_score_
rf_pred=rf_grid.predict(x_test)
print(f"Classification Report:{classification_report(y_test,rf_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,rf_pred)}")
print(f"Accuracy Score:{accuracy_score(y_test,rf_pred)}")
print(f"Best Score:",best_score_rf)
print(f"Best Params:",best_params_rf)
```

```python
rf_pred_proba=rf_grid.predict_proba(x_test)[:,1]
precision,recall,thresholds=precision_recall_curve(y_test,rf_pred_proba)
fpr,tpr,thresholds=roc_curve(y_test,rf_pred_proba)
auc=roc_auc_score(y_test,rf_pred_proba)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,6))
axes[0].plot(recall,precision,label='Random Forest Classifier')
axes[0].set_title('Precision-Recall Curve')
axes[0].set_xlabel('Recall')
axes[0].set_ylabel('precision')
axes[0].legend()
axes[1].plot(fpr,tpr,label=f'ROC Curve(AUC:{auc:.2f})',color='blue')
axes[1].set_title('Receiver Operating Characteristic Curve')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()
plt.tight_layout()
plt.show()
```
